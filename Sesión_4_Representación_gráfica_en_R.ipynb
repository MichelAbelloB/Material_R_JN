## 1 Representación gráfica en R

Se usa la función 


```
plot(x,y) # gráfico de la información numérica en el eje y contra el eje x
```
la cual tiene como argumentos principales son $x$ y $y$.



````r
x = c(2,5,7,-1,3)
y = c(3,2,-5,10,4)
plot(x,y)
````

Si no se incorpora el argumento y, R graficará a x como si estuviera en el eje vertical y en el eje horizontal irán los números desde 1 hasta la longitud del vector x:

````r
plot(x)  # En este caso se grafica x contra 1:length(x)
````

También se pueden graficar las funciones que uno usualmente ve en cálculo:

````r
f = function(x){log(x)}
plot(f)
````

## 1.1 Parámetros de la función *plot()*
### Parámetros fundamentales
1. *log*: para indicar que queremos el gráfico en escala logarítmica.
2. *type*: para elegir el estilo de línea que llevará el dibujo.
     
     "p" para puntos,

     "l" para líneas,

     "b" para ambas,

     "o" ambas sobrepuestas,
     
     "s" para escalones,

     "h" líneas parecidas a histograma,
      
     "n" para no hacer el dibujo de puntos.

3. *main("título")*: para colocar el título al gráfico. Si en lugar de texto se desea usar una expresión matemática, usamos  la función *expression()*.
4. *xlab("etiqueta del eje x")*: pone etiqueta al eje horizontal.
5. *ylab("etiqueta del eje y")*: pone etiqueta al eje vertical.
6. *pch=n* para elegir el símbolo de los puntos, $n=0,1,...,25$. El valor por defecto es *pch=1*.
7. "bg": para colocar un color de relleno si $21\leq pch \leq 25$.
8. *cex:* para elegir el tamaño de los símbolos.
9. *col="color en inglés"*: para elegir el color de los símbolos.

Se pueden consultar una paleta amplia de colores consultando el link:

## Ejemplo 1
Vamos a realizar el dibujo de la función exponencial $f(x)=2^x$:

````r
exponencial = function(x){exp(x)}
x = seq(0,5,0.1)
options(repr.plot.width=12,repr.plot.height=8)  # opciones para poner las imágenes más grandes
par(mfrow=c(1,2))                               # opción para poder establecer que se van a colocar dos dibujos en una matriz de una fila y dos columnas
plot(x,exponencial(x))                          # gráfico normal
plot(x,exponencial(x),log="y")                  # gráfico con escala logarítmica en el eje y 
par(mfrow=c(1,1))                               # opción para colocar los dibujos en una matriz de una fila por una columna   
````

````r
plot(x,exponencial(x),
     main = "Función exponencial",       
     xlab="eje x",ylab="eje y",
     pch = 22,bg='red',
     cex=2,
     col='darkcyan')
````

## 1.2 Diferentes tipos de gráficos


````r
par(mfrow = c(3,2))
x = seq(-10,10)
y = c(3,4,2,7,1,8,4,7,9,2,5,3,5,5,9,2,6,10,5,3,1)
plot(x,y,pch=23,cex=2,col="blue",bg="red",type="p",main="gr 1")
plot(x,y,pch=23,cex=2,col="cadetblue",bg="red",type="l",main="gr 2")
plot(x,y,pch=23,cex=2,col="chartreuse1",bg="red",type="b",main="gr 3")
plot(x,y,pch=23,cex=2,col="red",bg="red",type="o",main="gr 4")
plot(x,y,pch=23,cex=2,col="mediumpurple4",bg="red",type="h",main="gr 5")
plot(x,y,pch=23,cex=2,col="black",bg="red",type="s",main="gr 6")
par(mfrow = c(1,1))
````

##1.3 Más parámetros de la función *plot()*

* *lty* para cuando se trabaja con líneas se puede especificar los siguientes tipos:

   *  "dashed" para tipo de línea interrumpida,
   *  "solid" para tipo de línea continua,
   *  "dotted" para tipo de línea punteada y 
   *  "dotdashed" para tipo de línea que alterna rayas y puntos.
* *lwd* para especificar el grosor de la línea
* *xlim* para especificar el intervalo en el eje $x$ a graficar.
* *ylim* para especificar el intervalo en el eje $y$ a graficar.
* *xaxp* para modificar posiciones de las marcas en el eje $x$.
* *yaxp* para modificar posiciones de las marcas en el eje $y$.

### Ejemplo

````r
x = seq(-5,5,0.1)
y = -3*x^3-2*x^2+5*x+7
plot(x,y,
type="o",                          #Tipo
main="y=-3x^3-2x^2+5x+7 ",         #Título
pch=21,                            #estilo de punto número 21
bg="yellow",                       #relleno del punto
cex=1.8,                           #tamaño del punto 180%
lty="dashed",                      #estilo de línea punteada
lwd=2,                             #tamaño de línea
xlim=c(-5,-2),                     #Intervalo de visualización en el eje x
ylim=c(0,300),                     #Intervalo de visualización en el eje y
xaxp=c(-5,-2,3),                   #los primeros dos números son el intervalo de visualización en x, y 3 corresponde al número de etiquetas en x.
yaxp=c(0,300,10))                  #los primeros dos números son el intervalo de visualización en y, y 3 corresponde al número de etiquetas en y.
````

## 1.4 Adición de puntos y líneas
El comando
* *points(x,y)* adhiere al gráfico un vector de puntos en las coordenadas $x$ y   $y$.
* *abline* adhiere una recta a un gráfico de las siguientes formas
  
  * *abline(b,a)* coloca una recta del tipo $y=ax+b$,
  * *abline(v=x0)* coloca una recta vertical con ecuación $x=x_0$,
  * *abline(h=y0)* coloca una recta horizontal con ecuación $y=y_0$.

````r
x = sample(1:15)
y = sample(1:15)
linear_reg = lm(y~x) # linear model regresión lineal de y contra x
b = linear_reg$coeff[1]
a = linear_reg$coef[2]
plot(x,y)
abline(b,a,col='red')
````

## 1.5 Adición de texto
* *text(x,y,labels="...")* añade en el punto de coordenadas (x,y) el texto que se especifíca en labels.
  * *pos*: permite indicar la posición de texto alrededor de las coordenadas (x,y). Admite las entradas:
    
    * 1 para abajo
    * 2 para izquierda
    * 3 arriba
    * 4 derecha
    * sin especificar, el texto se sitúa centrado en el punto (x,y) 

````r
plot(x,y)
points(4,8,pch=23,bg="red")
text(4,8,labels="Aquí hay texto",pos=4)
````

##1.6 Adición de más gráficos y leyendas

Añadimos un gráfico de líneas a otro gráfico ya existente con las opciones

* *lines(x,y)*, aquí se añade un alínea poligonal que une los puntos $(x_i,y_i)$ de los vectores $x$ y $y$ de forma sucesiva.
* *curve(curva)* que permite añadir una curva a un gráfico existente.
   
   * *add=TRUE*, si no, la curva no se añade
   * la curva se puede especificar mediante una expresión algebraica con variable $x$, o mediante su nombre si se ha definido antes.

Añadimos leyenda a varias curvas usando la función

* *legend(posición,legend="...")*

    La posicion indica dónde queremos colocar la leyenda, se especifican mediante las posiciones en inglés:
    * "bottom" / "bottomright" / "bottomleft"
    * "top" / "topleft" / "topright"
    * "center" / "right" / "left"
  
  *legend: contiene el vector de nombres entre comillas con los que queremos identificar a las curvas en la leyenda

````r
x = seq(1,100,0.1)
y = log(x)
plot(x,y,type="l",col="black",ylim=c(-2,10))
lines(x,sqrt(x),col="blue")
lines(x,1/x,col="red")
legend("topleft",col=c("black","blue","red"),legend=c("log(x)","raíz de x","1/x"),lwd=3,bty="l")
````
